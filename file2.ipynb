```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv("car.csv")
df.head()
```

```python
df = pd.read_csv("car.csv")
x = df.iloc[:, :-1].values
y = df.iloc[: , -1].values
pd.DataFrame(x)
```

```python
print(x.shape)
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan , strategy='mean')
imputer.fit(x[ : , 7:11])
x[:, 7:11] = imputer.transform(x[:, 7:11])
pd.DataFrame(x[ 0:17 , 7:11])
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(),[0,3,4,5,6])],remainder='passthrough')
x = np.array(ct.fit_transform(x))
pd.DataFrame(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
print("Trainig Data..")
print("The shape of training features: ", x_train.shape)
print("The shape of training labels: ", y_train.shape)
```

```python
print("Trainig Data..")
print("The shape of training features: ", x_test.shape)
print("The shape of training labels: ", y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lin = LinearRegression()
lin.fit(x_train,y_train)
```

```python
train_accuracy = np.ceil(lin.score(x_train,y_train)*100)
print("The Training Accuracy = " ,train_accuracy,"%")
```

```python
y_predict = lin.predict(x_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_predict.reshape(len(y_predict),1), y_test.reshape(len(y_test),1)),1))
```

```python
train_accuracy = np.ceil(lin.score(x_test,y_test)*100)
print("The Testing Accuracy = " ,train_accuracy,"%")
```
