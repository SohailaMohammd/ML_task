```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
print("Trainig Data..")
print("The shape of training features: ", x_train.shape)
print("The shape of training labels: ", y_train.shape)
```

```python
print("Trainig Data..")
print("The shape of testing features: ", x_test.shape)
print("The shape of testing labels: ", y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
logistic = LogisticRegression()
logistic.fit(x_train,y_train)
```

```python
train_accuracy = np.ceil(logistic.score(x_train,y_train)*100)
print("The Training Accuracy = " ,train_accuracy,"%")
```

```python
train_accuracy = np.ceil(logistic.score(x_test,y_test)*100)
print("The Testing Accuracy = " ,train_accuracy,"%")
```
