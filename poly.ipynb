```python
import numpy as np
import pandas as pd
import matplotlib as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("car.csv")
df.head()
```

```python
df = pd.read_csv("car.csv")
x = df.iloc[:, :-1].values
y = df.iloc[: , -1].values
pd.DataFrame(x)
```

```python
print(x.shape)
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan , strategy='mean')
imputer.fit(x[ : , 7:11])
x[:, 7:11] = imputer.transform(x[:, 7:11])
pd.DataFrame(x[ 0:17 , 7:11])
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(),[0,3,4,5,6])],remainder='passthrough')
x = np.array(ct.fit_transform(x))
pd.DataFrame(x)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
poly_fea = PolynomialFeatures(degree=2)
x_train_poly = poly_fea.fit_transform(x_train)
x_test_poly = poly_fea.fit_transform(x_test)
```

```python
poly_reg = LinearRegression()
poly_reg.fit(x_train_poly,y_train)
```

```python
lin = LinearRegression()
lin.fit(x_train_poly,y_train)
```

```python
y_pred = poly_reg.predict(x_test_poly)
```

```python
train_accuracy = np.ceil(lin.score(x_train_poly,y_train)*100)
print("The Training Accuracy = " ,train_accuracy,"%")
```

```python
train_accuracy = np.ceil(lin.score(x_test_poly,y_test)*100)
print("The Testing Accuracy = " ,train_accuracy,"%")
```
